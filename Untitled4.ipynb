**A confusion matrix is a table used to evaluate** the performance of a classification model. It compares the predicted labels with the actual labels and provides a comprehensive summary of how well the model is performing. Here's how it works:

Structure of a Confusion Matrix
For a binary classification problem, the confusion matrix is a 2x2 table:

**Actual/Predicted	      Positive  	Negative**
Positive	True Positive (TP)	 False Negativ(FN)
Negative	False Positive (FP) 	True Negative (TN)

**True Positive (TP): The model correctly predicted** **the positive class.**

**True Negative (TN): The model correctly predicted** **the negative class**.

**False Positive (FP): The model incorrectly** **predicted the positive class (Type I error).**

**False Negative (FN): The model incorrectly** **predicted the negative class (Type II error).**

**Class 0: Negative**

**Class 1: Positive**

The confusion matrix will look like this:

                    Class 1 (Positive)	Class 0 (Negative)

Class 1 (Positive)	True Positive (TP)	False Negative (FN)

Class 0 (Negative)	False Positive (FP)	True Negative (TN)

# **BUILDING A PREDICTIVE MODEL**

Training on dataset 1: You train a logistic regression model with dataset 1 and assign it to model_lr.
Training on dataset 2: You then train a logistic regression model with dataset 2 and assign it again to model_lr.
Predictive model: You want to use model_lr for prediction.
